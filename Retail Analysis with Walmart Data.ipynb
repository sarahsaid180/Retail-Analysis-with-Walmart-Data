# Data Visualization

## 1. Temperature

```python
bins=[-2.06,20,40,60,80,100.14]
labels =['Less Than 20','From 20 To 40','From 40 To 60','From 60 To 80','More Than 80']
df['Temperature Category'] = pd.cut(df['Temperature'],bins,labels =labels)
df.head()
```

```python
df['Temperature Category'].value_counts().plot(kind = 'pie')
```

from the above chart we conclude that the most common / repeated degree is between 60 and 80 C

## The Relation between Weekly Sales and Temperature

```python
x = df['Temperature Category']
y = df['Weekly_Sales']
sns.catplot(data=df , x="Temperature Category", y="Weekly_Sales")
```

From the above Chart we can conclude that :
1. the Sales was at it's top when the temperature ranges between 20 and 60
2. the lowest Sales scored when the temperature is less than 20 C

## 2. Fuel Price

```python
bins=[2.472,3.5,4.468]
labels =['From 2.4 to 3.5','From 3.5 To 4.4']
df['Fuel Price Category'] = pd.cut(df['Fuel_Price'],bins,labels =labels)
df.head()
```

```python
fuel=df[["Fuel_Price"]]
fuel.boxplot()
plt.show()
```

from the above plot we can conclude that :
1. The lowest fuel price scored is approximately 2.4
2. The highest fuel price scored is approximately 4.4
3. The mean fuel price scored is approximately 3.4

## The relation between Weekly Sales and the Fuel Price

```python
x = df['Fuel Price Category']
y = df['Weekly_Sales']
sns.catplot(data=df , x="Fuel Price Category", y="Weekly_Sales")
```

from the above plot we can conclude that :
1. The Sales was at it's top we the fuel price was less than 3.5 , We can say that the highest sales is when the fuel price is less than the mean which is 2.4 .

2. The highest fuel price , The lowest sales.

# 3. CPI (Prevailing consumer price index)


```python
fuel=df[["CPI"]]
fuel.boxplot()
plt.show()
```

From the above the plot we can conclude :
1. The Min CPI scored is approximately 125

## The relation between Weekly Sales and CPI


```python
plt.figure(figsize=(8,6))
sns.regplot(x = df["CPI"], y = df["Weekly_Sales"])
plt.show()
```

from the above plot we can conclude that :
1. The Majority of sales have a CPI greater than 180 , In other words Most of the sales have a CPI greater than the mean which is 185
